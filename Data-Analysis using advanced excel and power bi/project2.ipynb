**VISUALIZATION**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (12,6))
sns.histplot(data = df, x = 'Total', bins = 30, kde = True)
plt.title('Total Stats')
plt.xlabel('Total Stats')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (12,6))
sns.scatterplot(data = df, x = 'Attack', y = 'Defense', hue = 'Legendary')
plt.title('Attack vs Defense')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.legend(title = 'Legendary', loc = 'upper right')
plt.show()
```

```python
plt.figure(figsize = (12,6))
sns.boxplot(data = df, x = 'Generation', y = 'Total', palette = 'coolwarm')
plt.title('Total Stats Distribution by Generation')
plt.xlabel('Generation')
plt.ylabel('Total Stats')
plt.show()
```

```python
legendary_counts = df['Legendary'].value_counts()
plt.figure(figsize = (8,8))
plt.pie(legendary_counts, labels = legendary_counts.index, autopct='%1.1f%%', colors=['skyblue','lightcoral'])
plt.title('Proportion of Legendary Pokemon')
plt.show()
```

```python
print(legendary_counts)
```

```python
type1_c = df['Type 1'].value_counts()
plt.figure(figsize = (10,8))
plt.pie(type1_c, labels = type1_c.index, startangle = 90,counterclock = False,autopct='%1.1f%%')
plt.title('Distribution of Type 1 Pokemon')
plt.axis('equal')
plt.show()
```

```python
#Corelation Matrix
numerical_columns = df[['Total','HP','Attack','Defense','Sp. Atk','Sp. Def','Speed','Generation']]
corr_matrix = numerical_columns.corr()
plt.figure(figsize = (12,8))
sns.heatmap(corr_matrix, annot=True,cmap='coolwarm',linewidths=.5)
plt.title('Correlation of POkemon stats')
plt.show()
```

```python
#3D Scatter Plot
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = (10,8))
ax = fig.add_subplot(111,projection = '3d')
x = df['Attack']
y = df['Defense']
z = df['Total']
colors = df['Generation']
sc = ax.scatter(x,y,z, c = colors, cmap =  'viridis', s=50)
ax.set_xlabel('Attack')
ax.set_xlabel('Defense')
ax.set_xlabel('Total')
ax.set_title('3D Scatter Plot')
plt.colorbar(sc, label = 'Generation')
plt.show()
```

```python
sns.pairplot(df[['Total','HP','Attack','Defense','Sp. Atk','Sp. Def','Speed','Generation','Legendary']],hue = 'Legendary', palette = 'husl')
plt.suptitle('Pair of Plot of PokemON')
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.violinplot(x = 'Legendary' , y = 'Speed',data = df, inner='quartile',palette='muted')
plt.title('Violin Plot of Speed by Legendary')
plt.xlabel('Legendary')
plt.ylabel('Speed')
plt.xticks(ticks=[0,1],labels=['Non Legendary','Legendary'])
plt.show()
```

**FEATURE ENGINEERING**

```python
 #NEW FEATURE
df['Total_without_HP'] = df['Total'] - df['HP']
```

```python
df.head()
```

```python
#Normalizing feature
from sklearn.preprocessing import MinMaxScaler
#Normalizing attack,defense,speed
scaler = MinMaxScaler()
df[['Attack_scaled','Defence_scaled','Speed_scaled']] = scaler.fit_transform(df[['Attack','Defense','Speed'] ])
```

```python
df.head()
```

```python
#binning
bins = [0,50,100,150,200]
labels = ['Very Slow','Slow','Fast','Very Fast']
df['Speed_cat'] = pd.cut(df['Speed'],bins = bins, labels = labels)
```

```python
df.head()
```

```python
#label Encoding
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Legendary_transform'] = le.fit_transform(df['Legendary'])
```
